```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/priya/Documents/Final Capstone/Large Files/Cleaned_Cusine_Data.csv', low_memory = False)  
```

```python
plt.style.use('ggplot')
```

```python
#full dataframe
plt.figure(figsize=(11,7))
grouped = df.groupby('category')['review_count'].count().sort_values(ascending = False)
sns.barplot(y=grouped.index, x= grouped.values, palette= sns.color_palette("Blues", len(grouped)) )
for i,v in enumerate(grouped):
    plt.text(v, i+0.15, str(v), fontsize=14)
plt.tick_params(labelsize=14)
```

```python
#full dataframe
#Average length of words in each category
table = df.groupby(['category','labels'])['num_words_review'].mean().round().unstack()
plt.figure(figsize=(11,8))
sns.heatmap(table, cmap='Blues', fmt='g',annot=True, linewidths=1)
plt.tick_params(labelsize=15)
```

```python
#bigdata
plt.figure(figsize=(11,7))
grouped = df.review_stars.value_counts().sort_index()
sns.barplot(grouped.index, grouped.values, palette=sns.color_palette("Blues", len(grouped)))
plt.tick_params(labelsize=14)
for  i, v in enumerate(grouped):
    plt.text(i, v*1.02, str(v), horizontalalignment ='center',fontweight='bold', fontsize=14)
```

```python
grouped = df.groupby('category')['review_stars'].mean().round(2).sort_values(ascending=False)
sns.pointplot(grouped.index, grouped.values)
plt.ylim(3)
for  i, v in enumerate(grouped):
    plt.text(i, v, str(v), horizontalalignment ='center',fontweight='bold', fontsize=14)    
```
